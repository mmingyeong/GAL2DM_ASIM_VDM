```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

path = "/gpfs/adupuy/cosmicweb_asim/ASIM_TSC/samples/training/0.hdf5"
bins = 200

# zoom 
P_LO, P_HI = 1, 99          # xlim 1~99  
MIN_WIDTH_FACTOR = 1e-2     #   global std 1%    
EPS = 1e-12

with h5py.File(path, "r") as f:
    vpec = f["input"][0][1].astype(np.float64)

def standardize(x, eps=1e-12):
    mu = np.mean(x)
    sig = np.std(x)
    sig = sig if sig > eps else eps
    return (x - mu) / sig, mu, sig

def safe_log10_x_plus_1(x, eps=1e-12):
    xmin = np.min(x)
    shift = -xmin if xmin < 0 else 0.0
    y = np.log10((x + shift) + 1.0 + eps)
    return y, shift

def robust_xlim(x, p_lo=1, p_hi=99, min_width_factor=1e-2, eps=1e-12):
    """xlim percentile , lo~hi      ."""
    xr = x.ravel()
    lo, hi = np.percentile(xr, [p_lo, p_hi])
    mu = np.mean(xr)
    sig = np.std(xr)
    sig = sig if sig > eps else eps

    width = hi - lo
    min_width = sig * min_width_factor

    # lo~hi   ,     
    if (not np.isfinite(width)) or width < min_width:
        half = max(min_width / 2, 1e-6)  #  0 
        lo, hi = mu - half, mu + half

    #  lo==hi  
    if abs(hi - lo) < eps:
        lo, hi = mu - 1.0, mu + 1.0

    return lo, hi

# ===== 3 cases =====
v1, shift = safe_log10_x_plus_1(vpec)
v2, mu2, sig2 = standardize(vpec)
v3, mu3, sig3 = standardize(v1)

# xlim  ()
xlim1 = robust_xlim(v1, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS)
xlim2 = robust_xlim(v2, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS)
xlim3 = robust_xlim(v3, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS)

# ===== Plot =====
fig, axes = plt.subplots(1, 3, figsize=(18, 5), constrained_layout=True)

axes[0].hist(v1.ravel(), bins=bins)
axes[0].set_xlim(*xlim1)
axes[0].set_title(f"Case 1) safe log10(x+1)\nshift={shift:.4g}, xlim={xlim1[0]:.4g}..{xlim1[1]:.4g}")
axes[0].set_xlabel("value"); axes[0].set_ylabel("count")

axes[1].hist(v2.ravel(), bins=bins)
axes[1].set_xlim(*xlim2)
axes[1].set_title(f"Case 2) Standardize\nxlim={xlim2[0]:.4g}..{xlim2[1]:.4g}")
axes[1].set_xlabel("value")

axes[2].hist(v3.ravel(), bins=bins)
axes[2].set_xlim(*xlim3)
axes[2].set_title(f"Case 3) log10(x+1)+Std\nxlim={xlim3[0]:.4g}..{xlim3[1]:.4g}")
axes[2].set_xlabel("value")

plt.show()

# ()     
def quick_stats(name, x):
    xr = x.ravel()
    q = np.percentile(xr, [0, 1, 50, 99, 100])
    print(f"{name}: min={q[0]:.6g}, p1={q[1]:.6g}, median={q[2]:.6g}, p99={q[3]:.6g}, max={q[4]:.6g}, std={np.std(xr):.6g}")

quick_stats("vpec(raw)", vpec)
quick_stats("case1(log)", v1)
quick_stats("case2(std)", v2)
quick_stats("case3(log+std)", v3)
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# =========================
# Config
# =========================
path = "/gpfs/adupuy/cosmicweb_asim/ASIM_TSC/samples/training/0.hdf5"
bins = 200
EPS = 1e-12

# For display zoom (x-axis), without dropping data
P_LO, P_HI = 0.5, 99.5
MIN_WIDTH_FACTOR = 1e-2  # if p-range too narrow, use std*factor as minimum width

# =========================
# Load vpec
# =========================
with h5py.File(path, "r") as f:
    vpec = f["input"][0][1].astype(np.float64)

v_all = vpec.ravel()
mask_nz = (v_all != 0)           # IMPORTANT: mask from RAW data
v_nz = v_all[mask_nz]
zero_frac = 1.0 - (v_nz.size / v_all.size)

# =========================
# Helpers
# =========================
def standardize(x, eps=1e-12):
    mu = np.mean(x)
    sig = np.std(x)
    sig = sig if sig > eps else eps
    return (x - mu) / sig, mu, sig

def signed_log10_1p_abs(x, eps=1e-12):
    # y = sign(x) * log10(1 + |x|)
    return np.sign(x) * np.log10(1.0 + np.abs(x) + eps)

def robust_xlim(x, p_lo=0.5, p_hi=99.5, min_width_factor=1e-2, eps=1e-12):
    """xlim from percentiles; if too narrow, enforce minimum width around mean."""
    xr = np.asarray(x).ravel()
    lo, hi = np.percentile(xr, [p_lo, p_hi])
    mu = np.mean(xr)
    sig = np.std(xr)
    sig = sig if sig > eps else eps

    width = hi - lo
    min_width = sig * min_width_factor

    if (not np.isfinite(width)) or width < min_width:
        half = max(min_width / 2, 1e-6)
        lo, hi = mu - half, mu + half

    if abs(hi - lo) < eps:
        lo, hi = mu - 1.0, mu + 1.0

    return lo, hi

def stats_str(x):
    xr = np.asarray(x).ravel()
    return (
        f"min={xr.min():.3g}\n"
        f"max={xr.max():.3g}\n"
        f"mean={xr.mean():.3g}\n"
        f"std={xr.std():.3g}"
    )

def add_box(ax, text, loc=(0.98, 0.98)):
    ax.text(
        loc[0], loc[1], text,
        transform=ax.transAxes,
        ha="right", va="top",
        fontsize=9,
        bbox=dict(facecolor="white", alpha=0.75, edgecolor="none")
    )

# =========================
# Build 4 cases
# =========================
# Case 1: signed log (raw -> signed log)
case1_all = signed_log10_1p_abs(v_all)
case1_nz  = case1_all[mask_nz]
case1_formula = r"$y=\mathrm{sign}(x)\log_{10}(1+|x|)$"

# Case 2: standardize (raw -> zscore)
case2_all, mu2, sig2 = standardize(v_all)
case2_nz = case2_all[mask_nz]
case2_formula = r"$y=(x-\mu)/\sigma$"

# Case 3: signed log + standardize (raw -> signed log -> zscore)
case3_all, mu3, sig3 = standardize(case1_all)
case3_nz = case3_all[mask_nz]
case3_formula = r"$y=\frac{\mathrm{sign}(x)\log_{10}(1+|x|)-\mu}{\sigma}$"

# Case 4: standardize + signed log (raw -> zscore -> signed log)
case4_all = signed_log10_1p_abs(case2_all)
case4_nz  = case4_all[mask_nz]
case4_formula = r"$y=\mathrm{sign}(z)\log_{10}(1+|z|),\; z=(x-\mu)/\sigma$"

cases_all = [case1_all, case2_all, case3_all, case4_all]
cases_nz  = [case1_nz,  case2_nz,  case3_nz,  case4_nz]
formulas  = [case1_formula, case2_formula, case3_formula, case4_formula]
titles    = [
    "Case 1) signed log10(1+|x|)",
    "Case 2) standardize",
    "Case 3) signed log -> std",
    "Case 4) std -> signed log"
]

# xlims (separately for each subplot row/case)
xlims_all = [robust_xlim(c, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS) for c in cases_all]
xlims_nz  = [robust_xlim(c, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS) for c in cases_nz]

# =========================
# Plot 2x4: (row0=ALL), (row1=RAW-nonzero-mask)
# =========================
fig, axes = plt.subplots(2, 4, figsize=(26, 9), constrained_layout=True)

fig.suptitle(
    f"VPEC Normalization Comparison (ALL vs. Raw-Nonzero Mask)\n"
    f"raw_zero_frac={zero_frac*100:.2f}%  |  raw_shape={vpec.shape}  dtype=float64",
    fontsize=14
)

for j in range(4):
    # ---- Row 0: ALL ----
    ax = axes[0, j]
    ax.hist(cases_all[j], bins=bins, alpha=0.85)
    ax.set_xlim(*xlims_all[j])
    ax.set_title(f"{titles[j]}  (ALL)")
    ax.set_xlabel("value")
    if j == 0:
        ax.set_ylabel("count")

    box0 = (
        formulas[j] + "\n\n" +
        "[ALL stats]\n" + stats_str(cases_all[j]) + "\n\n" +
        f"xlim(p{P_LO}..p{P_HI})\n"
        f"{xlims_all[j][0]:.3g} .. {xlims_all[j][1]:.3g}"
    )
    add_box(ax, box0)

    # ---- Row 1: RAW nonzero mask ----
    ax = axes[1, j]
    ax.hist(cases_nz[j], bins=bins, alpha=0.85)
    ax.set_xlim(*xlims_nz[j])
    ax.set_title(f"{titles[j]}  (RAW nonzero mask)")
    ax.set_xlabel("value")
    if j == 0:
        ax.set_ylabel("count")

    box1 = (
        formulas[j] + "\n\n" +
        "[RAW-nonzero stats]\n" + stats_str(cases_nz[j]) + "\n\n" +
        f"raw_zero_frac={zero_frac*100:.2f}%\n"
        f"xlim(p{P_LO}..p{P_HI})\n"
        f"{xlims_nz[j][0]:.3g} .. {xlims_nz[j][1]:.3g}"
    )
    add_box(ax, box1)

plt.show()
```

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# VPEC Normalization: Case 1 ~ 10 (single file, integrated)
#   Rows layout (4 x 5):
#     Row 1: ALL (includes zeros) for Case 1-5
#     Row 2: RAW nonzero only     for Case 1-5
#     Row 3: ALL (includes zeros) for Case 6-10
#     Row 4: RAW nonzero only     for Case 6-10
# ============================================================

# =========================
# Config
# =========================
path = "/gpfs/adupuy/cosmicweb_asim/ASIM_TSC/samples/training/0.hdf5"
bins = 200
EPS = 1e-12

# display zoom (x-axis): percentile window
P_LO, P_HI = 0.5, 99.5
MIN_WIDTH_FACTOR = 1e-2

# robust percentiles for Case 7/8, 9/10
ROBUST_P_LO, ROBUST_P_HI = 0.5, 99.5
CLIP_ROBUST = True

# =========================
# Helpers
# =========================
def standardize(x, eps=1e-12):
    mu = float(np.mean(x))
    sig = float(np.std(x))
    sig = sig if sig > eps else eps
    return (x - mu) / sig, mu, sig

def signed_log10_1p_abs(x, eps=1e-12):
    return np.sign(x) * np.log10(1.0 + np.abs(x) + eps)

def minmax_scale(x, xmin, xmax, lo=0.0, hi=1.0, eps=1e-12):
    denom = (xmax - xmin)
    denom = denom if denom > eps else eps
    y01 = (x - xmin) / denom
    return y01 * (hi - lo) + lo

def robust_xlim(x, p_lo=0.5, p_hi=99.5, min_width_factor=1e-2, eps=1e-12):
    xr = np.asarray(x).ravel()
    lo, hi = np.percentile(xr, [p_lo, p_hi])
    mu = np.mean(xr)
    sig = np.std(xr)
    sig = sig if sig > eps else eps

    width = hi - lo
    min_width = sig * min_width_factor

    if (not np.isfinite(width)) or width < min_width:
        half = max(min_width / 2, 1e-6)
        lo, hi = mu - half, mu + half

    if abs(hi - lo) < eps:
        lo, hi = mu - 1.0, mu + 1.0

    return lo, hi

def stats_str(x):
    xr = np.asarray(x).ravel()
    return (
        f"n={xr.size:,}\n"
        f"min={xr.min():.3g}\n"
        f"max={xr.max():.3g}\n"
        f"mean={xr.mean():.3g}\n"
        f"std={xr.std():.3g}"
    )

def add_box(ax, text, loc=(0.98, 0.98)):
    ax.text(
        loc[0], loc[1], text,
        transform=ax.transAxes,
        ha="right", va="top",
        fontsize=8,
        bbox=dict(facecolor="white", alpha=0.75, edgecolor="none")
    )

# =========================
# Load vpec
# =========================
with h5py.File(path, "r") as f:
    vpec = f["input"][0][1].astype(np.float64)

v_all = vpec.ravel()
mask_nz = (v_all != 0)  # RAW nonzero mask
v_nz = v_all[mask_nz]
zero_frac = 1.0 - (v_nz.size / v_all.size)

if v_nz.size == 0:
    raise RuntimeError("No nonzero values found (all zeros). Cannot compute some cases.")

# =========================
# Precompute stats
# =========================
xmin_raw, xmax_raw = float(np.min(v_all)), float(np.max(v_all))
mu_raw, sig_raw = float(np.mean(v_all)), float(np.std(v_all))
sig_raw = sig_raw if sig_raw > EPS else EPS

log_raw = signed_log10_1p_abs(v_all, EPS)
mu_log, sig_log = float(np.mean(log_raw)), float(np.std(log_raw))
sig_log = sig_log if sig_log > EPS else EPS

# robust percentiles:
# - For Case 7/8 (apply-to-all), we compute pL/pH on RAW nonzero (recommended for sparse)
pL_nz, pH_nz = np.percentile(v_nz, [ROBUST_P_LO, ROBUST_P_HI])
if (not np.isfinite(pL_nz)) or (not np.isfinite(pH_nz)) or abs(pH_nz - pL_nz) < 1e-12:
    pL_nz, pH_nz = float(np.min(v_nz)), float(np.max(v_nz))

# =========================
# Build Case 1 ~ 10 (ALL arrays)
# =========================
# Case 1: signed log
case1_all = log_raw

# Case 2: standardize raw
case2_all = (v_all - mu_raw) / sig_raw

# Case 3: signed log -> standardize
case3_all = (case1_all - mu_log) / sig_log

# Case 4: standardize -> signed log
case4_all = signed_log10_1p_abs(case2_all, EPS)

# Case 5: min-max [0,1] using raw min/max
case5_all = minmax_scale(v_all, xmin_raw, xmax_raw, 0.0, 1.0, EPS)

# Case 6: min-max [-1,1] using raw min/max
case6_all = minmax_scale(v_all, xmin_raw, xmax_raw, -1.0, 1.0, EPS)

# Case 7: robust min-max [0,1] apply-to-all, robust range from RAW nonzero
v_base7 = np.clip(v_all, pL_nz, pH_nz) if CLIP_ROBUST else v_all
case7_all = minmax_scale(v_base7, pL_nz, pH_nz, 0.0, 1.0, EPS)

# Case 8: robust min-max [-1,1] apply-to-all
case8_all = minmax_scale(v_base7, pL_nz, pH_nz, -1.0, 1.0, EPS)

# Case 9: robust min-max [0,1] on NONZERO only (zeros stay 0)
case9_all = np.zeros_like(v_all, dtype=np.float64)
nz_vals = v_all[mask_nz]
if CLIP_ROBUST:
    nz_vals = np.clip(nz_vals, pL_nz, pH_nz)
case9_all[mask_nz] = minmax_scale(nz_vals, pL_nz, pH_nz, 0.0, 1.0, EPS)

# Case 10: robust min-max [-1,1] on NONZERO only (zeros stay 0)
case10_all = np.zeros_like(v_all, dtype=np.float64)
case10_all[mask_nz] = minmax_scale(nz_vals, pL_nz, pH_nz, -1.0, 1.0, EPS)

cases_all = [
    case1_all, case2_all, case3_all, case4_all, case5_all,
    case6_all, case7_all, case8_all, case9_all, case10_all,
]
cases_nz = [c[mask_nz] for c in cases_all]  # RAW nonzero only view (mask fixed from raw)

titles = [
    "Case 1) signed log10(1+|x|)",
    "Case 2) standardize",
    "Case 3) signed log -> std",
    "Case 4) std -> signed log",
    "Case 5) min-max [0,1] (raw min/max)",
    "Case 6) min-max [-1,1] (raw min/max)",
    f"Case 7) robust min-max [0,1] (p{ROBUST_P_LO}..p{ROBUST_P_HI}, clip={CLIP_ROBUST})",
    f"Case 8) robust min-max [-1,1] (p{ROBUST_P_LO}..p{ROBUST_P_HI}, clip={CLIP_ROBUST})",
    f"Case 9) nonzero-only robust [0,1] (zeros stay 0, clip={CLIP_ROBUST})",
    f"Case 10) nonzero-only robust [-1,1] (zeros stay 0, clip={CLIP_ROBUST})",
]
formulas = [
    r"$y=\mathrm{sign}(x)\log_{10}(1+|x|)$",
    r"$y=(x-\mu)/\sigma$",
    r"$y=\frac{\mathrm{sign}(x)\log_{10}(1+|x|)-\mu}{\sigma}$",
    r"$y=\mathrm{sign}(z)\log_{10}(1+|z|),\; z=(x-\mu)/\sigma$",
    r"$y=\frac{x-x_{\min}}{x_{\max}-x_{\min}}$",
    r"$y=2\frac{x-x_{\min}}{x_{\max}-x_{\min}}-1$",
    r"$y=\frac{\mathrm{clip}(x,p_L,p_H)-p_L}{p_H-p_L}$",
    r"$y=2\frac{\mathrm{clip}(x,p_L,p_H)-p_L}{p_H-p_L}-1$",
    r"$y=\mathbf{1}_{x\neq0}\cdot\frac{\mathrm{clip}(x,p_L,p_H)-p_L}{p_H-p_L}$",
    r"$y=\mathbf{1}_{x\neq0}\cdot\left(2\frac{\mathrm{clip}(x,p_L,p_H)-p_L}{p_H-p_L}-1\right)$",
]

# xlims per plot (ALL vs NZ)
xlims_all = [robust_xlim(c, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS) for c in cases_all]
xlims_nz  = [robust_xlim(c, P_LO, P_HI, MIN_WIDTH_FACTOR, EPS) for c in cases_nz]

# =========================
# Plot: 4 x 5
# =========================
fig, axes = plt.subplots(4, 5, figsize=(28, 18), constrained_layout=True)
fig.suptitle(
    f"VPEC Normalization Comparison (Case 1~10)  single file\n"
    f"raw_shape={vpec.shape} | raw_zero_frac={zero_frac*100:.2f}% | "
    f"robust on RAW-nonzero: p{ROBUST_P_LO}={pL_nz:.3g}, p{ROBUST_P_HI}={pH_nz:.3g} | clip={CLIP_ROBUST}",
    fontsize=14
)

def plot_hist(ax, data, xlim, title, formula, extra):
    ax.hist(data, bins=bins, alpha=0.85)
    ax.set_yscale("log")
    ax.set_xlim(*xlim)
    ax.set_title(title, fontsize=10)
    ax.set_xlabel("value")
    ax.grid(True, alpha=0.2)
    add_box(ax, f"{formula}\n\n{extra}")

# Row mapping:
# row0: ALL case 1-5
# row1: NZ  case 1-5
# row2: ALL case 6-10
# row3: NZ  case 6-10
for j in range(5):
    # case index for first block
    idx1 = j
    plot_hist(
        axes[0, j], cases_all[idx1], xlims_all[idx1],
        f"{titles[idx1]}\n(ALL / includes zeros)", formulas[idx1],
        "[ALL stats]\n" + stats_str(cases_all[idx1]) + f"\n\nxlim(p{P_LO}..p{P_HI})\n{xlims_all[idx1][0]:.3g}..{xlims_all[idx1][1]:.3g}"
    )
    plot_hist(
        axes[1, j], cases_nz[idx1], xlims_nz[idx1],
        f"{titles[idx1]}\n(RAW nonzero only)", formulas[idx1],
        "[RAW-nonzero stats]\n" + stats_str(cases_nz[idx1]) +
        f"\n\nraw_zero_frac={zero_frac*100:.2f}%\n"
        f"xlim(p{P_LO}..p{P_HI})\n{xlims_nz[idx1][0]:.3g}..{xlims_nz[idx1][1]:.3g}"
    )

    # case index for second block
    idx2 = 5 + j
    plot_hist(
        axes[2, j], cases_all[idx2], xlims_all[idx2],
        f"{titles[idx2]}\n(ALL / includes zeros)", formulas[idx2],
        "[ALL stats]\n" + stats_str(cases_all[idx2]) + f"\n\nxlim(p{P_LO}..p{P_HI})\n{xlims_all[idx2][0]:.3g}..{xlims_all[idx2][1]:.3g}"
    )
    plot_hist(
        axes[3, j], cases_nz[idx2], xlims_nz[idx2],
        f"{titles[idx2]}\n(RAW nonzero only)", formulas[idx2],
        "[RAW-nonzero stats]\n" + stats_str(cases_nz[idx2]) +
        f"\n\nraw_zero_frac={zero_frac*100:.2f}%\n"
        f"xlim(p{P_LO}..p{P_HI})\n{xlims_nz[idx2][0]:.3g}..{xlims_nz[idx2][1]:.3g}"
    )

# y-label only on leftmost column
for r in range(4):
    axes[r, 0].set_ylabel("count (log)")

plt.show()
```
